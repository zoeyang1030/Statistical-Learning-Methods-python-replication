### Monte-Carlo Cross Validation

```python
model_list = ['traditional', 'dualform', 'pocket', 'voted', 'averaged', 'margin']
name_list = ['Perceptron Algorithm', 'Perceptron Dual-form', 'Pocket Algorithm', 
             'Voted Perceptron', 'Averaged Perceptron', 'Perceptron with Margin']
name_dict = dict(zip(model_list, name_list))
score_dict = {'Model': list(), 'Training Score': list(), 'Testing Score': list(), 'Time': list()}
for model_name in model_list:
    tr_score = np.array([])
    te_score = np.array([])
    time_score = np.array([])
    for i in range(100):
        X_tr, X_te, y_tr, y_te = split_train_test(X, y, 0.5)
        model = PerceptronLinearAlgorithm(model_type=model_name, max_iter=1e4, lr=0.02, margin=0.01)
        start = time.perf_counter()
        model.fit(X_tr, y_tr)
        time_score = np.append(time_score, time.perf_counter() - start)
        tr_score = np.append(tr_score, model.score(X_tr, y_tr))
        te_score = np.append(te_score, model.score(X_te, y_te))
    score_dict['Model'].append(name_dict[model_name])
    score_dict['Training Score'].append(round(tr_score.mean(), 3))
    score_dict['Testing Score'].append(round(te_score.mean(), 3))
    score_dict['Time'].append(round(time_score.mean(), 3))

pd.DataFrame(score_dict)
```
