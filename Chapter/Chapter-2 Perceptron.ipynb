```python
import numpy as np
import matplotlib.pyplot as plt

from Tools.sample import *
from Tools.visualization import *
from Models.Perceptron import *
```

### Generate & Split Dataset

```python
X, y = gen_class(n=80, interval=1, label_type=1, random_seed=1)
X_train, X_test, y_train, y_test = split_train_test(X, y, 0.5)
data_viz = LinearClassifierPlot(title='Generated Sample')
data_viz.scatter_plot(X, y)
```

### The Original Perceptron Algorithm

```python
perceptron = PerceptronLinearAlgorithm(model_type='traditional', max_iter=1e4, lr=0.01)
perceptron.fit(X_train, y_train)
print('Training Score: %s\nTesting Score: %s'
      %(round(perceptron.score(X_train, y_train), 2), 
        round(perceptron.score(X_test, y_test), 2)))

fig = plt.figure(figsize=(13.5,5))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

data_viz = LinearClassifierPlot(ax=ax1, title='Generated Sample')
data_viz.scatter_plot(X, y)
perceptron_tr = LinearClassifierPlot(ax=ax2, title='Perceptron Algorithm (Train)')
perceptron_tr.scatter_plot(X_test, y_test)
perceptron_tr.hyperplane_plot(perceptron)
perceptron_te = LinearClassifierPlot(ax=ax3, title='Perceptron Algorithm (Test)')
perceptron_te.scatter_plot(X_test, y_test)
perceptron_te.hyperplane_plot(perceptron)
```
